```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.pyplot import figure
import scipy.io
from obspy import read, read_inventory, Stream
from obspy.core import UTCDateTime as UT
from obspy.geodetics.base import gps2dist_azimuth
from obspy.signal.rotate import rotate_zne_lqt
from obspy.clients.fdsn import Client
from scipy.io import savemat

hawai = Client('IRIS')
start = UT('2018-05-18')
end   = start+86400
h_stations = hawai.get_stations(network='*',station='*',location='*',channel='HN?',
                                starttime=start,endtime=end,minlatitude=19.3262,
                                maxlatitude=19.47,minlongitude=-155.35,maxlongitude=-155.1443)[0]  
h_inv = hawai.get_stations(network='*',station='*',location='*',channel='HN?',
                                starttime=start,endtime=end,minlatitude=19.3262,
                                maxlatitude=19.47,minlongitude=-155.35,maxlongitude=-155.1443, level = 'response')                   
station_names = [station.code for station in h_stations]
station_latlon = [[station.latitude, station.longitude] for station in h_stations]
print(station_names)
print(station_latlon)
print(h_inv)
h_inv[1].plot_response(min_freq=0.001);
h_inv[0][0][0].response

h_bd_stations = hawai.get_stations(network='*',station='*',location='*',channel='HH*',
                                starttime=start,endtime=end,minlatitude=18.85,
                                maxlatitude=20.37,minlongitude=-156.13,maxlongitude=-154.79)[0] 
h_bd_inv = hawai.get_stations(network='*',station='*',location='*',channel='HH*',
                                starttime=start,endtime=end,minlatitude=18.85,
                                maxlatitude=20.37,minlongitude=-156.13,maxlongitude=-154.79, level = 'response') 

station_bd_names = [station.code for station in h_bd_stations]
station_bd_latlon = [[station.latitude, station.longitude] for station in h_bd_stations]
h_bd_inv[1].plot_response(min_freq=0.001);

print(station_bd_names)
print(station_bd_latlon)
print(h_bd_inv)

#Define functions to retrieve and plot data
def get_acce_data(time,length, station_names):
  """
  function get_acce_data
  retrives acceleration data for the station list given above, for a given time and a given length 
  input time: obspy.core.UTCDateTime
  param time: reference time to cut the waveforms
  input length: float
  param length: number of seconds to get the waveforms
  """
  hst = Stream()
  start = time
  end = start + length
  for station_name in station_names:
      try:
          hst += hawai.get_waveforms(network='HV',station=station_name,location='*',channel='HN*',starttime=start,endtime=end-0.01)
      except Exception as e:
          print(e) 
  return hst


def get_bd_data(time,length, station_names):
  """
  function get_bd_data
  retrives broadband data for the station list given above, for a given time and a given length 
  input time: obspy.core.UTCDateTime
  param time: reference time to cut the waveforms
  input length: float
  param length: number of seconds to get the waveforms
  """
  hst = Stream()
  start = time
  end = start + length
  for station_name in station_names:
      try:
          hst += hawai.get_waveforms(network='*',station=station_name,location='*',channel='HH*',starttime=start,endtime=end-0.01)
      except Exception as e:
          print(e) 
  return hst

def intT(tr):
    # I want to integrate from as close to the onset of the event as possible, and integrate for 60s.
    tstart = tr.times("utcdatetime")[0]+50;
    tend = tstart + 60;
    trc = tr.copy().slice(starttime=tstart, endtime=tend, nearest_sample=True);
    
    # First detrend the acceleration
    trc.detrend("linear");
    
    # Then manually remove the mean of the first few seconds of the data
    trc_s = trc.copy().slice(starttime=tstart-0.1, endtime=tstart, nearest_sample=False);
    mean_s = trc_s.copy().data.mean(); 
    
    trc.data = trc.data - mean_s;
    
    # Deconvolution of instrument response
    try:
        trc = trc.copy().remove_response(inventory=h_inv, output="ACC",
                  water_level=10, plot=False)
    except:
        # Exception is raised for HMLE, because the waveform starttime is outside of the response time range
        true_starttime = trc.copy().stats.starttime # true start time of HMLE waveform
        
        trc.stats.location = ''
        
        trc.stats.starttime = UT('2018-05-04') # change it to an arbitrary time within response time range
        
        
        trc = trc.copy().remove_response(inventory=h_inv, output="ACC",
                  water_level=10, plot=False) # remove response
        
        trc.stats.starttime = true_starttime # change back to the true start time of HMLE waveform
   
    
    # Correct for static shift again after deconvolution
    trc_s = trc.copy().slice(starttime=tstart-0.1, endtime=tstart, nearest_sample=True);
    mean_s = trc_s.copy().data.mean(); 
    trc.data = trc.data - mean_s;
    acc = trc;
    
    # Integrate in time to get velocity 
    vel = trc.copy().integrate(method='cumtrapz'); 
    
    # Detrend the velocity using linear model
    vel.detrend("linear"); 
    
    # Correct for static shift in velocity before integrating to displacement
    vel_s = vel.copy().slice(starttime=tstart-0.1, endtime=tstart, nearest_sample=True);
    meanV_s = vel_s.copy().data.mean(); 
    vel.data = vel.data - meanV_s;
    
    # Integrate in time to get displacement
    disp = vel.copy().integrate(method='cumtrapz'); # displacement
    
    # Detrend the displacement using 3rd order spline
    #disp.detrend("spline", order=3,dspline=500); 
    
    r = [acc, vel, disp];
    
    return r
```
